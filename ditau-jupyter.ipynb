```python
datafile_name = '06D59B66-2858-3E42-B867-8FDE59F4D206.root'

if not os.path.isfile(datafile_name):
    ! curl -O http://opendata.cern.ch/eos/opendata/cms/Run2016G/Tau/NANOAOD/UL2016_MiniAODv2_NanoAODv9-v1/270000/06D59B66-2858-3E42-B867-8FDE59F4D206.root
```

```python
infile = uproot.open(datafile_name)
events = infile['Events']
df = events.arrays(filter_name='/(run|event|luminosityBlock|nTau|Tau_)/', library='pd')
```

```python
df = df[df['nTau'] == 2]
```

```python
df
```

```python
df.drop('nTau', inplace=True, axis=1)
```

```python
df.drop('run', inplace=True, axis=1)
df.drop('event', inplace=True, axis=1)
df.drop('luminosityBlock', inplace=True, axis=1)
```

```python
df.columns
```

```python
new_df = pd.DataFrame()

for c in df.columns:

    ndf = pd.DataFrame(
        df[c].tolist(),
        columns=[f'{c}_0', f'{c}_1']
    )

    new_df = pd.concat([new_df,ndf], axis=1)
```

```python
new_df.to_csv(
    'ditau.csv',
    index=False,
)
```
